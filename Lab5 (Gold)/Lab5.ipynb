# TÌM GIÁ VÀNG CAO NHẤT TRONG VIỆT NAM 6 THÁNG TRƯỚC

```python
import requests
from bs4 import BeautifulSoup
import json
import re
```

```python
path = "https://webgia.com/gia-vang/sjc/bieu-do-6-thang.html"
```

```python
request = requests.get(path)
soup = BeautifulSoup(request.text,'html.parser')
```

```python
#Define java script code ,get data from the last script 
javacode = soup.find_all('script')[-1].text
```

```python
# Define patterns to match the "Bán ra" and "Mua vào" series data
ban_ra_pattern = re.compile(r'name:"Bán ra",data:\[\[.*?]],')
mua_vao_pattern = re.compile(r'name:"Mua vào",data:\[\[.*?]],')

# Find and extract the data for "Bán ra" and "Mua vào" series
ban_ra_matches = re.findall(ban_ra_pattern, javacode)
mua_vao_matches = re.findall(mua_vao_pattern, javacode)
```

```python
#Put data in the different category
ban_ra_data = []
mua_vao_data = []

# Process "Bán ra" data
for match in ban_ra_matches:
    data_points = re.findall(r'\[(.*?)]', match)
    for data_point in data_points[1:]:
        ban_ra_data.append(data_point)

# Process "Mua vào" data
for match in mua_vao_matches:
    data_points = re.findall(r'\[(.*?)]', match)
    for data_point in data_points[1:]:
        mua_vao_data.append(data_point)

# Print or process the data as needed
print("Bán ra data:")
for data_point in ban_ra_data:
    print([data_point[14:19]])

print("\nMua vào data:")
for data_point in mua_vao_data:
    print([data_point[14:19]])
```

```python
#Extract the value
ban_ra_prices = [data[14:19] for data in ban_ra_data]
mua_vao_prices = [data[14:19] for data in mua_vao_data]

# Find the highest values
highest_ban_ra = max(ban_ra_prices)
highest_mua_vao = max(mua_vao_prices)

# Print the highest values
print(f'Giá vàng bán ra cao nhất của Việt Nam: {highest_ban_ra} triệu đồng/lượng')
print(f'Giá vàng mua vào cao nhất của Việt Nam: {highest_mua_vao} triệu đồng/lượng')
```
